```python
import streamlit as st

# Set page title
st.set_page_config(page_title="Urooj Moin's Resume", layout="wide")

# Sidebar Navigation
st.sidebar.title("Navigation")
selection = st.sidebar.radio("Go to", ["Home", "Projects", "Skills", "Contact"])

# Header
st.title("📜 Urooj Moin's Resume")

# Contact Information
st.header("📞 Contact Information")
st.write("📧 **Email:** [uroojmoin9@gmail.com](mailto:uroojmoin9@gmail.com)")
st.write("📞 **Phone:** 03378202012")
st.write("🔗 **LinkedIn:** [linkedin.com/in/UroojMoin](https://linkedin.com/in/UroojMoin)")
st.write("💻 **GitHub:** [github.com/UroojMoin](https://github.com/UroojMoin)")

# About Me Section
if selection == "Home":
    st.header("👩‍💻 About Me")
    st.write(
        "I am a self-driven and motivated student at **Bahria University Karachi**, "
        "studying **Artificial Intelligence**. Passionate about AI and always eager to learn new skills."
    )

    # Education Section
    st.header("🎓 Education")
    st.write("**B.S. in Artificial Intelligence** | Bahria University Karachi | 2025")

# Projects Section
elif selection == "Projects":
    st.header("🚀 Projects")
    projects = {
        "Created a webpage of electronics using HTML.": "Basic webpage using HTML & CSS.",
        "Student Report Card Management System using C++.": "Desktop-based student report system.",
        "Blood Bank Management System using Java and GUI.": "Java GUI application for managing blood donations.",
        "Blood Donation Database using SQL.": "SQL-based system for blood donor records.",
        "SOHO using Cisco.": "Network security implementation using Cisco.",
        "Secret Message Encryption Decryption using Python.": "Python-based cryptography project.",
        "Prediction of Genre from Netflix Meta Data using ML concepts.": "Machine Learning project for genre classification.",
        "Emotion Detection System Images Dataset using CV concepts.": "Computer Vision project for facial emotion recognition.",
    }

    for project, description in projects.items():
        with st.expander(f"📌 {project}"):
            st.write(description)

# Skills Section
elif selection == "Skills":
    st.header("💡 Skills & Certifications")
    skills = [
        "C++", "Java", "Python", "HTML", "SQL", "Cisco",
        "Machine Learning", "Computer Vision", "NLP"
    ]

    for skill in skills:
        st.markdown(f"✅ **{skill}**")

    st.subheader("📜 Certification")
    st.write("🏅 Mobile App Development")

# Contact Section
elif selection == "Contact":
    st.header("📩 Get in Touch")
    st.write("Feel free to reach out to me via email or LinkedIn.")

    # Contact Form
    with st.form("contact_form"):
        name = st.text_input("Name")
        email = st.text_input("Email")
        message = st.text_area("Message")
        submit_button = st.form_submit_button("Send")

        if submit_button:
            st.success("✅ Message sent successfully!")

# Download Resume as PDF
st.sidebar.download_button(
    label="📄 Download Resume",
    data="Resume PDF Data Here",  # Replace with actual PDF file data
    file_name="Urooj_Moin_Resume.pdf",
    mime="application/pdf"
)
```
